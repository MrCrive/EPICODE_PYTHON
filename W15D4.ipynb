# DATAFRAME GUADAGNI MENSILI DIVERSE ANNATE

```python
import numpy as np
```

```python
years = 5
a = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years),
                            "Anno": np.repeat(list(range(years)), 12),
                            "Guadagni": np.random.randint(800, 5000, 12*years)})
```

```python
a
```

```python
a.Mese = list( ["GENNAIO","FEBBRAIO","MARZO","APRILE","MAGGIO","GIUGNO",
                    "LUGLIO","AGOSTO","SETTEMBRE","OTTOBRE","NOVEMBRE","DICEMBRE"]*years)
```

```python
a.Mese
```

```python
a.Guadagni.cumsum()
```

```python
anni = a.Anno.unique()
for anno in anni:
    print("Anno:",anno)
    filtro_anno = a.Anno == anno
    somma_cumulativa_annuale = a.loc[filtro_anno, "Guadagni"].cumsum()
    print(somma_cumulativa_annuale)
    print("-"*10)
```
